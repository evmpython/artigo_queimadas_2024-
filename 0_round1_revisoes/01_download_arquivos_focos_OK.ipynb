# Função de download dos dados

```python
def download_focos(url, dir_output, filename):
    url_filename = f'{url}{filename}'
    myfile = requests.get(url_filename, verify=False)
    open(f'{dir_output}/{filename}', 'wb').write(myfile.content)
    return
```

# Download dos dados

```python
# ano inicial de final
anoi, anof = '2003','2025'

# ftp dos dados anual do satélite AQUA
url = f'https://dataserver-coids.inpe.br/queimadas/queimadas/focos/csv/anual/Brasil_sat_ref/'

# loop dos meses
ano_mesi = f'{anoi}0101'
ano_mesf = f'{anof}1201'

for data in pd.date_range(ano_mesi, ano_mesf, freq='1Y'):

    # estrai ano e mês
    ano = data.strftime('%Y')
    mes = data.strftime('%m')

    # nome do arquivo
    filename = f'focos_br_ref_{ano}.zip' # Exemplo: focos_br_ref_2003.zip

    print('BAIXAND ===>>>', filename)

    # baixa o arquivo
    download_focos(url, dir_output, filename)
```
